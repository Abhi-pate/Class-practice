```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, output_file,show
from bokeh.sampledata.iris import flowers
output_file('test.html')
p= figure(title='trst_flower')
p.xaxis.axis_label= 'patel_lenght'
p.yaxis.axis_label='patel_width'
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
from bokeh.plotting import figure, output_file,show
from bokeh.sampledata.iris import flowers
output_file('test.html')
p= figure(title='line_graph')
p.xaxis.axis_label= 'patel_lenght'
p.yaxis.axis_label='patel_width'
p.line(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
x=[3,4,5,6,7]
y=[11,2,3,4,5]
output_file('line.html')
p= figure(title='line_graph')
p.xaxis.axis_label= 'x-axis'
p.yaxis.axis_label='y-axis'
p.line(x,y)
show(p)
```

```python
x=[3,4,5,6,7]
y=[11,2,3,4,5]
output_file('line.html')
p= figure(title='line_graph')
p.xaxis.axis_label= 'x-axis'
p.yaxis.axis_label='y-axis'
p.scatter(x,y)
show(p)
```

```python
x=[3,4,5,6,7]
y=[11,2,3,4,5]
output_file('line.html')
p= figure(title='line_graph')
p.xaxis.axis_label= 'x-axis'
p.yaxis.axis_label='y-axis'
p.line(x,y,line_color='red')
show(p)
```

```python
x=[3,4,5,6,7]
y=[11,2,3,4,5]
output_file('line.html')
p= figure(title='line_graph')
p.xaxis.axis_label= 'x-axis'
p.yaxis.axis_label='y-axis'
p.scatter(x,y,fill_color='red')
show(p)
```

```python

x=[3,4,5,6,7]
y=[11,2,3,4,5]
output_file('line.html')
p= figure(title='line_graph')
p.xaxis.axis_label= 'x-axis'
p.yaxis.axis_label='y-axis'
p.scatter(x,y,fill_color='red',legend_label='red point')
show(p)
```

```python

x=[3,4,5,6,7]
y=[11,2,3,4,5]
output_file('line.html')
p= figure(title='line_graph')
p.xaxis.axis_label= 'x-axis'
p.yaxis.axis_label='y-axis'
p.scatter(x,y,fill_color='red',legend_label='red point',size=20)
show(p)
```
